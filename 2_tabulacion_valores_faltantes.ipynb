Observando la tabla, podriamos agruparla según las variables a las cueles tienen valores faltantes

```python
retail.missing.missing_variable_table()
```

# Casos / Observaciones / Filas

```python
#Resumenes por caso
retail.missing.missing_case_summary()
```

```python
retail.missing.missing_case_table()
```

## intervalos de valores faltamtes
Es el momentgo de salir de la caja, debemos realizar más preguntas

```python
(
    retail
    .missing
    .missing_variable_span(
        variable = "Description",
        span_every = 50
    )
)
```

# Run length de valores faltantes

```python
(
    retail
    .missing
    .missing_variable_run(
        variable = "Description",
    )
)
```
