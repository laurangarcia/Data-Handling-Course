```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
retail = pd.read_csv('online_retail.csv')
```

```python

retail.missing.missing_variable_plot()
```

## Casos / Observaciones / Filas

```python
retail.missing.missing_case_plot()
```

(

    retail

    .missing

    .missing_variable_span_plot(

        variable = "Description",

        span_every = 50000,

        rot = 0
        
    )
)

De esta forma la funcin est intentando crear un arreglo extremadamente grande, que no cabe en la memoria RAM de la computadora.

Esto suele pasar porque:

* El DataFrame es muy grande.
* El parmetro span_every genera demasiados intervalos.
* La funcin no est optimizada y crea estructuras de datos innecesariamente grandes.

```python
retail.missing.missing_variable_span_plot(
    variable="Description",
    span_every=5000,  # Ajusta segn tamao de tus datos
    rot=45
)
```

```python
#diagrama de barras para visualizar los datos faltantes
msno.bar(df=retail)
```

```python
msno.matrix(df=retail)

#Una barra totalmente negra tiene todos los datos, con lineas blancas son los faltantes
```
